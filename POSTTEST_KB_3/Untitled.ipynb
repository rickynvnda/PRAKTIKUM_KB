<h2>GRAFIK BATANG</h2>
<h4>Grafik ini biasa digunakan untuk membandingkan kategori. Kita dapat memunculkan visualisasi data dengan menggunakan syntax berikut. (x) digunakan untuk menanmpilkan Tipe Pekerjaan, (y) digunakan untuk menampilkan umur</h4>

```python
x = df['work_type']
y = df['age']
```

```python
plt.bar(x, y, align="edge", width=0.5,)
plt.title("Healthcare Dataset")
plt.xlabel("Tipe Pekerjaan")
plt.ylabel("Umur")
plt.show()
```

```python
df.dtypes
```

<h2>GRAFIK SCATTER PLOT</h2>
<h4>diagram grafis yang dibangun dari dua sumbu X dan Y (variabel X dan variabel Y). Scatter plot menunjukan hubungan antara dua variabel. </h4>

```python
sns.scatterplot(data = df, x='work_type', y='age')
plt.title("Persebaran Data Healthcare")
plt.show()
```

<h2>GRAFIK BOXPLOT</h2>
<h4>grafik ini digunakan untuk melihat persebaran data dan mencari outliers. (x) digunakan untuk menanmpilkan Status Merokok, (y) digunakan untuk menampilkan Umur</h4>

```python
sns.boxplot(x= df['smoking_status'], y=df['age'])
plt.title("Healthcare Dataset")
plt.show()
```

<h2>GRAFIK PIE/PIE CHART</h2>
<h4>Grafik Pie cocok digunakan pada data categorical. Grafik ini biasa digunakan untuk melihat distribusi.</h4>

```python
df['gender'].value_counts()
```

```python
labels = df['smoking_status'].value_counts().rename_axis('smoking_status').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title("Healthcare Dataset")
plt.pie(df['smoking_status'].value_counts(), labels=labels['smoking_status'], autopct='%.1f%%', startangle=90)
plt.show()
```

<h2>GRAFIK HEATMAP</h2>
<h4>Grafik ini biasa digunakan untuk menunjukkan korelasi antar variable, dengan cara menampilkan data dengan cara representasi warna yang berbeda-beda dan semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap.</h4>

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), cmap='Oranges')
ax.set_title("Healthcare Dataset")
plt.show()
```
